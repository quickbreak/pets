# 1.3.4

image.png

```python
import numpy as np

n = 3
m = 3
a = np.array([[0]*m]*n)
print(f'firstly:\n{a}')
for i in range(m):
  a[0][i] = 1
  a[n - 1][i] = 1

print(f'secondly:\n{a}')

for i in range(n):
  a[i][0] = 1
  a[i][m - 1] = 1

print(f'by the end:\n{a}')
```

```python
import pandas as pd
from sklearn import preprocessing as p

link = 'https://raw.githubusercontent.com/akmand/datasets/master/iris.csv'

a = pd.read_csv(link, on_bad_lines='skip')

a[['sepal_length_cm']] = p.MinMaxScaler(feature_range=(0, 1)).fit_transform(a[['sepal_length_cm']])
a[['sepal_width_cm']] = p.StandardScaler().fit_transform(a[['sepal_width_cm']])
a
```
