```python
import numpy as np

x = int(input('Первое число: '))
y = int(input('Второе число: '))
#operation = input("Действие: '+', '-', '*', '/', 'e**(x+y)','sin(x+y)', 'cos(x+y)', или 'x**y' ")

operation = int(input('Действия:\n1. x+y\n2. x-y\n3. x*y\n4. x/y\n5. e**(x+y)\n6. sin(x+y)\n7. cos(x+y)\n8. x**y\nНомер: '))
match operation:
    case 1:
        print(f'{x} + {y} = {x+y}')
    case 2:
        print(f'{x} - {y} = {x-y}')
    case 3:
        print(f'{x} * {y} = {x*y}')
    case 4:
        try:
            print(f'{x} / {y} = {x/y}')
        except ZeroDivisionError:
            print(f'Неверный ввод. На ноль делить нельзя')
    case 5:
        print(f'e**({x} + {y}) = {np.exp(x+y)}')
    case 6:
        print(f'sin({x} + {y}) = {np.sin(x+y)}')
    case 7:
        print(f'cos({x} + {y}) = {np.cos(x+y)}')
    case 8:
        print(f'{x}**{y} = {x**y}')
```
